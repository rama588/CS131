```python
from pyspark.sql import SparkSession
import time
import matplotlib.pyplot as plt
```

```python
import time

start = time.time()

df_437 = spark.read.option("header", "true").option("inferSchema", "true").csv("gs://dataproc-staging-us-central1-920508693171-4om4tbcd/artificial_data_files_ws78/data_437rows.csv")
df_437.groupBy("passenger_count").avg("fare_amount").show()

end = time.time()
print(f"Runtime for data_437rows.csv (54 kb): {end - start:.3f} seconds")
```

```python
kb_sizes = [54, 112.6, 220.1, 384.6, 737.7]
times = [1.198, 1.138, 0.988, 1.480, 1.169]
labels = ["54 KB", "112.6 KB", "220.1 KB", "384.6 KB", "737.7 KB"]

plt.figure(figsize=(8, 5))
plt.plot(kb_sizes, times, marker='o', linestyle='-', color='blue')
plt.title("Runtime vs. File Size")
plt.xlabel("File Size (KB)")
plt.ylabel("Runtime (seconds)")
plt.grid(True)

for i in range(len(kb_sizes)):
    plt.text(kb_sizes[i] + 10, times[i], labels[i], fontsize=9)

plt.tight_layout()
plt.show()
```
